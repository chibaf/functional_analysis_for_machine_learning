リスト 5.1

```python
import numpy as np, matplotlib.pyplot as plt, math
from scipy.integrate import odeint

def firstOrderSystem(x, t, k, vc):
    dxdt = - k * (x - vc)
    return dxdt

t = np.arange(0.0, 10, 0.01)
v0, vc = 0, 1
k = 1 
v = odeint(firstOrderSystem, v0, t, args = (k, vc))

fig, ax = plt.subplots()
ax.plot(t, vc*np.ones(len(t)), ls='--'), ax.plot(t, v)
plt.xlabel('Time [s]'), plt.ylabel('$v_c, v$ [m/s]'), plt.show()
```

リスト 5.2

```python
t = np.arange(0.0, 15, 0.01)
p0, pd = 0.5, 1.5
k = 0.4 
p = odeint(firstOrderSystem, p0, t, args = (k, pd))

fig, ax = plt.subplots()
ax.plot(t, pd*np.ones(len(t)), ls='--'), ax.plot(t, p)
plt.xlabel('Time [s]'), plt.ylabel('$p$ [m]'), plt.show()
```

リスト 5.3

```python
def trackingSystem(p, t, k):
  dpdt = 2*np.pi/5*np.cos(2*np.pi/5*t) \
        - k * (p - np.sin(2*np.pi/5*t))
  return dpdt

t = np.arange(0.0, 15, 0.01)
p0, pd = 0.0, np.sin(2*math.pi/5*t) 
k = 0.4 
p = odeint(trackingSystem, p0, t, args = (k,))

fig, ax = plt.subplots()
ax.plot(t, pd), ax.plot(t, p)
plt.xlabel('Time [s]'), plt.ylabel('$p$ [m]'), plt.show()
```
